```python
#importing data
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_Poland = df[df['job_country'] == 'Poland'].copy()
```

```python
df_Poland
```

```python
df_Poland['job_posted_month'] = df_Poland['job_posted_date'].dt.strftime('%B')

df_Poland
```

```python
df_Poland_pivot = df_Poland.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_Poland_pivot.reset_index(inplace=True)

df_Poland_pivot['month_no'] = pd.to_datetime(df_Poland_pivot['job_posted_month'], format='%B').dt.month

df_Poland_pivot.sort_values('month_no', inplace=True)
df_Poland_pivot.set_index('job_posted_month', inplace=True)
df_Poland_pivot.drop(columns='month_no', inplace=True)
```

```python
df_Poland_pivot
```

```python
df_Poland_pivot.plot(kind='line')
```

```python
top_3 = df_Poland['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
 
```

```python
df_Poland_pivot[top_3].plot(kind='line')

plt.title('Monthly Job Postings Fot the top Data jobs in Poland')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
df_Poland_software_pivot = pd.read_csv('http://lukeb.co/software_csv', index_col='job_posted_month')
```

```python
df_Poland_merged = df_Poland_pivot.merge(df_Poland_software_pivot, on="job_posted_month")
```

```python
top_5 = df_Poland_merged.sum().sort_values(ascending=False).head().index.to_list()
top_5
```

```python
df_Poland_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings For Top Tech Jobs in Poland ')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
